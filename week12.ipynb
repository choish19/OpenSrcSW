```python
#실습문제 1

import csv

f = open('seoul.csv','r',encoding='cp949')
data = csv.reader(f, delimiter=',')
header = next(data)
max_temp = -999
max_date = ''
min_temp = 999
min_date = ''
max_dif = 0
max_dif_date = ''
dif = 0
for row in data:
    if row[-1] == '':
        row[-1] = -999
        dif = -1
    row[-1] = float(row[-1])
    if max_temp < row[-1]:
        max_date = row[0]
        max_temp = row[-1]
    if row[-2] == '':
        row[-2] = 999
        dif = -1
    row[-2] = float(row[-2])
    if min_temp > row[-2]:
        min_date = row[0]
        min_temp = row[-2]
    if dif != -1:
        dif = row[-1] - row[-2]
        if max_dif < dif:
            max_dif_date = row[0]
            max_dif = dif
print('일교차가 가장 큰 날')
print(max_dif_date + " " + str(max_temp))
print('최저기온이 가장 낮은 날')
print(min_date + " " + str(min_temp))
print('최고기온이 가장 높은 날')
print(max_date + " " + str(max_dif))
f.close()
```

```python
#실습문제 2

import csv

f = open('subway.csv','r',encoding='cp949')
data = csv.reader(f, delimiter=',')
header = next(data)
pre = '1호선'
mx = [-1]*3
mn = [999999999]*3
mx_station = ['']*4
mn_station = ['']*4
for row in data:
    if pre != row[1]:
        print(pre)
        print('이용객이 가장 많은 역 순위')
        for i in range(3):
            print(str(i+1)+'. '+mx_station[i]+'('+str(mx[i])+')')
        print('이용객이 가장 적은 역 순위')
        for i in range(3):
            print(str(i+1)+'. '+mn_station[i]+'('+str(mn[i])+')')
        mx = [-1]*3
        mn = [999999999]*3
        mx_station = ['']*4
        mn_station = ['']*4
    
    row[-1] = row[-1].replace(",","")
    row[-2] = row[-2].replace(",","")
    count = int(row[-1]) + int(row[-2])
    for i in range(0,3):
        if(mx[i] < count):
            for j in range(2,i,-1):
                mx[j] = mx[j-1]
                mx_station[j] = mx_station[j-1]
            mx[i] = count
            mx_station[i] = row[3] + ' ' + row[1]
            break
    for i in range(0,3):
        if(mn[i] > count):
            for j in range(2,i,-1):
                mn[j] = mn[j-1]
                mn_station[j] = mn_station[j-1]
            mn[i] = count
            mn_station[i] = row[3] + ' ' + row[1]
            break
    pre = row[1]
f.close()
```
